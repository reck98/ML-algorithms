```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path as FilePath
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

warnings.filterwarnings("ignore")

FILE_PATH = FilePath("cleaned_data.csv")
```

```python
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
X_test, X_train, y_test, y_train = train_test_split(
    df.drop("survived", axis=1),
    df["survived"],
    test_size=0.2,
    random_state=42,
)
```

```python
X_test.head()
```

```python
def modelEvaluation(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
    return y_pred


def modelTraining(X_train, y_train, X_test, y_test, n = 5):
    model = KNeighborsClassifier(n_neighbors=n)
    
    model.fit(X_train, y_train)
    return modelEvaluation(model, X_test, y_test)
```

```python
for i in [1, 3, 5, 7, 9]:
    modelTraining(X_test=X_test, y_test=y_test, X_train=X_train, y_train=y_train, n=i)
    print("\n")
```
